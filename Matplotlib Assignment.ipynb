```python
"""Matplotlib is a popular Python library for creating static, animated, and interactive visualizations in Python. It provides a wide range of tools and functions for generating high-quality plots, graphs, charts, and other visual representations of data."""

## Matplotlib is used for various purposes, including:

# 1.Data exploration and analysis: Matplotlib allows you to quickly visualize and analyze data, helping you understand patterns, trends, and relationships within your dataset.
# 2.Presentation and communication: Matplotlib enables you to create visually appealing plots and charts, making it easier to communicate your findings and insights to others.
# 3.Report generation: Matplotlib's plots can be integrated into reports and publications, allowing you to present your data in a clear and visually appealing manner.
# 4.Dashboard development: Matplotlib is often used in building interactive dashboards and data visualization applications.
# 5.Educational purposes: Matplotlib is widely used for teaching and learning purposes, as it provides a straightforward and intuitive interface for creating visualizations.

"""The Pyplot module of Matplotlib provides a simple and convenient API for creating various types of plots. Five commonly used plots that can be created using the Pyplot module are:"""

# 1.Line plot: Used to represent data points connected by straight lines. It is suitable for visualizing trends and changes over time.
# 2.Bar plot: Displays rectangular bars of varying heights, representing the frequency or magnitude of a particular category or variable. It is often used to compare different categories or groups.
# 3.Scatter plot: Shows the relationship between two variables by displaying individual data points as dots on a two-dimensional coordinate system. It is useful for identifying correlations or clusters in data.
# 4.Histogram: Represents the distribution of a continuous variable by dividing it into intervals (bins) and showing the frequency or count of values falling into each bin.
# 5.Pie chart: Represents data as slices of a circular pie, with each slice corresponding to a particular category or proportion of the whole. It is suitable for displaying proportions or percentages.
```

```python
# Ans - A scatter plot is a diagram where each value in the data set is represented by a dot.
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
plt.scatter(x,y)
plt.show()
```

```python
"""The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to plot multiple plots or graphs in separate axes within the same figure window, enabling you to compare and visualize multiple datasets or aspects of data side by side."""
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0,100,200,300,400,500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure and a grid of subplots with 2 rows and 2 columns
fig, axs = plt.subplots(2, 2)

# Plotting the first line plot in the first subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plotting the second line plot in the first subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plotting the third line plot in the first subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plotting the fourth line plot in the first subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Adjust the spacing between subplots
plt.tight_layout()

plt.show()
```

```python
"""A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution."""
```

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]

fig = plt.figure(figsize =(8, 5))

bp = plt.boxplot(data)

plt.show()
```
